```python
dfq1 = df.loc[ : , ["Customer_Age","Card_Category","Avg_Open_To_Buy"] ].groupby( ["Customer_Age","Card_Category"] , as_index = False ).quantile(0.25)
dfq1.columns = ["Customer_Age","Card_Category","prom_buy_q1"]
```

```python
dfq3 = df.loc[ : , ["Customer_Age","Card_Category","Avg_Open_To_Buy"] ].groupby( ["Customer_Age","Card_Category"] , as_index = False ).quantile(0.75)
dfq3.columns = ["Customer_Age","Card_Category","prom_buy_q3"]
```

```python
#Este usa la media o percentil 50
dfq2 = df.loc[ : , ["Customer_Age","Card_Category","Avg_Open_To_Buy"] ].groupby( ["Customer_Age","Card_Category"] , as_index = False ).quantile(0.5)
dfq2.columns = ["Customer_Age","Card_Category","prom_buy_q2"]
```

```python
##realizamos la union de nuestro dataframe respecto a los cuartiles
df_iqr = dfq1.merge( dfq3 , on=["Customer_Age","Card_Category"] , how="inner" ).merge( dfq2 , on=["Customer_Age","Card_Category"] , how="inner" )
```

```python
df_iqr.head(10)
```

```python
#Representamos el iqr
df_iqr["iqr"] = df_iqr.apply( lambda r: r["prom_buy_q3"] - r["prom_buy_q1"] , axis = 1)
```

```python
df_iqr.head(10)
```

```python
df_iqr["upper"] = df_iqr.apply( lambda r: ( r["prom_buy_q3"] + 1.5*r["iqr"] ) , axis = 1)
df_iqr["lower"] = df_iqr.apply( lambda r: ( r["prom_buy_q1"] - 1.5*r["iqr"] ) , axis = 1)
```

```python
df_iqr.head(2)
```

```python
#Tenemos los limites, estamos listos para reemplazar los valores atipicos
```

```python
df_t = df.merge( df_iqr , on=["Customer_Age","Card_Category"] , how="left" ).copy()
```

```python
df_t
```

```python
#Funcion para identificar los valores atipicos y reemplezarlos por la media.

def getRealPromedio(r):
    if ( r["Avg_Open_To_Buy"] <= r["lower"] ) | ( r["Avg_Open_To_Buy"] >= r["upper"] ):
        return r["prom_buy_q2"]
    else:
        return r["Avg_Open_To_Buy"]
```

```python
df_t["new_prom_buy"] = df_t.apply( lambda r: getRealPromedio(r) , axis = 1 )
```
